# Homework 8 Plots

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
pi = np.pi
c0 = 343 # Speed of sound in air (m/s)
cw = 1480 # Speed of sound in water (m/s)
c = c0 # Try observing differences in results in air vs. water
theta = np.arange(0.001, 2*pi, 0.001)
```

## Problem 4: Line Sources

```python
# Parameters
f = [100, 1000, 10000] # Test with any number of frequencies (Hz)
L = [5, 0.5, 0.05] # Test with any number of source lengths (m)

# Directivity Function
def D_line(f, L, theta):
    w = 2*pi*f
    k = w/c
    inp = 0.5*k*L*np.sin(theta)
    line = np.sin(inp)/inp
    return line

# Plots
fig, ax = plt.subplots(len(f), len(L), subplot_kw = dict(projection = 'polar'), figsize = (len(L)*5, (len(f)*5)))
for i in range(len(f)):
  for j in range(len(L)):
    ax[j, i].set_thetamin(-90)
    ax[j, i].set_thetamax(90)
    ax[j, i].set_theta_zero_location("N")
    ax[j, i].plot(theta, abs(D_line(f[i], L[j], theta)), 'k')
    title = 'L = ' + str(L[j]) + 'm, f = ' + str(f[i]) + ' Hz';
    ax[j, i].set_title(title)
fig.suptitle('Line Source Directivities')
```

## Problem 5: Pair of Microphones

```python
# Parameters
d = 5e-1 # Try modifying mic separation distance (m)
f = [100, 1000, 10000] # Test with any number of frequencies (Hz)
tau = [0, 0.2, 0.6, -0.6, 0.8] # Test with any number of tau values (s)

# Directivity Function
def D_mic(f, tau, theta):
    w = 2*pi*f
    k = w/c
    inp = k*d*(np.sin(theta) + tau)
    mic = 1 - np.exp(-1j*inp)
    return mic

# Plots
fig, ax = plt.subplots(len(f), len(tau), subplot_kw = dict(projection = 'polar'), figsize = (len(tau)*5, len(f)*5))
for i in range(len(tau)):
  for j in range(len(f)):
    ax[j, i].plot(theta, abs(D_mic(f[j], tau[i], theta)), 'k')
    title = 'tau = ' + str(tau[i]) + ', f = ' + str(f[j]) + ' Hz';
    ax[j, i].set_title(title)
fig.suptitle('2-Mic Directivities, Separation = ' + str(d) + ' m')
```
